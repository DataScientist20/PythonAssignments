Q6. Why are lists mutable, but tuples are immutable ?
- **Lists:**
Designed for storing collections of items that may need to be modified. Their mutability allows for efficient addition, removal, and modification of elements. This flexibility comes at the cost of slightly more memory overhead and potentially slower iteration compared to tuples.
- **Tuples:**
Primarily used for grouping related data together, especially when the data should not be changed. Immutability ensures data integrity and allows tuples to be used as keys in dictionaries and elements in sets, which require hashable objects. Tuples are more memory-efficient and can be faster to iterate through than lists.

Q4. Write a program that checks if a given word is a Python keyword.

````python
import keyword

def is_keyword(word):
    """Checks if a word is a Python keyword.
    Args:
        word: The word to check.
    Returns:
        True if the word is a keyword, False otherwise.
    """
    return keyword.iskeyword(word)

# Example usage
word1 = "for"
word2 = "async"
word3 = "hello"

print(f"'{word1}' is a keyword: {is_keyword(word1)}")
print(f"'{word3}' is a keyword: {is_keyword(word2)}")
````

Q6. Write a function to demonstrate the behavior of mutable and immutable arguments.
- Mutable Arguments
Mutable objects are those whose values can be changed after they are created. Examples of mutable data types in Python include: lists, dictionaries, and sets.
When a mutable object is passed as an argument to a function, the function receives a reference to the original object. If the function modifies the object, the changes will be visible outside the function as well.

````python
def modify_list(my_list):
    my_list.append(4)
    my_list[0] = 10

my_list = [1, 2, 3]
modify_list(my_list)
print(my_list)  # Output: [10, 2, 3, 4]
````

- Immutable objects are those whose values cannot be changed after they are created. Examples of immutable data types in Python include: Integers, Floats, Strings, and Tuples.
When an immutable object is passed as an argument to a function, the function receives a copy of the object's value. If the function modifies the copy, the original object will not be affected.

````python
def modify_string(string1):
    string1 += "dfg"
    string1 = "abc"

string1 = "abcd"
modify_string(string1)
print(string1)  # Output: abcd
````

Q9. Write a Python program to convert user input from string to integer, float, and boolean types

````python
def convert_input(user_input):
    # Initialize dictionary to store the conversions
    conversions = {
        'integer': None,
        'float': None,
        'boolean': None
    }

    # Try to convert to integer
    if user_input.isdigit() or (user_input[1:].isdigit() and user_input[0] in ['-', '+']):
        conversions['integer'] = int(user_input)

    # Try to convert to float
    try:
        conversions['float'] = float(user_input)
    except ValueError:
        pass  # If it can't be converted to float, leave it as None

    # Define possible boolean representations
    true_values = {'true', '1', 'yes', 'on'}
    false_values = {'false', '0', 'no', 'off'}

    # Try to convert to boolean
    if user_input.lower() in true_values:
        conversions['boolean'] = True
    elif user_input.lower() in false_values:
        conversions['boolean'] = False

    return conversions

if __name__ == "__main__":
    # Get user input
    user_input = input("Enter a value: ")

    # Convert the input
    result = convert_input(user_input)

    # Output the results
    print(f"Integer: {result['integer']}")
    print(f"Float: {result['float']}")
    print(f"Boolean: {result['boolean']}")
````

Q10.  Write code to demonstrate type casting with list elements

````python
# Original list with different types of elements
my_list = ["10", "20.5", "True", "False", "123", 5, 3.14, True, False]

# Creating a new list to store the type-casted elements
converted_list = []

for element in my_list:
    if isinstance(element, str):
        # If the element is a string, try converting it to int or float or boolean
        if element.isdigit():
            converted_list.append(int(element))  # Convert string to integer
        else:
            try:
                converted_list.append(float(element))  # Convert string to float
            except ValueError:
                # Handle cases like "True" or "False"
                if element.lower() == "true":
                    converted_list.append(True)
                elif element.lower() == "false":
                    converted_list.append(False)
                else:
                    converted_list.append(element)  # Keep the original value
    elif isinstance(element, int):
        # If the element is an integer, convert it to string and boolean
        converted_list.append(str(element))  # Convert int to string
        converted_list.append(bool(element))  # Convert int to boolean
    elif isinstance(element, float):
        # If the element is a float, convert it to int and string
        converted_list.append(int(element))  # Convert float to int (truncates decimal part)
        converted_list.append(str(element))  # Convert float to string
    elif isinstance(element, bool):
        # If the element is boolean, convert it to string and integer
        converted_list.append(str(element))  # Convert bool to string
        converted_list.append(int(element))  # Convert bool to integer (1 for True, 0 for False)

# Print the original and converted lists
print("Original List:", my_list)
print("Converted List:", converted_list)
````

Q14. Write a program to reverse a string using a while loop

````python
# Get user input
original_string = input("Enter a string: ")

# Reverse the string using slicing
reversed_string = original_string[::-1]

# Print the reversed string
print("Reversed String:", reversed_string)

# Initialize an empty string for the reversed result
reversed_string = ""

# Get the index of the last character
index = len(original_string) - 1

# Use a while loop to iterate backwards
while index >= 0:
    reversed_string += original_string[index]  # Append characters in reverse order
    index -= 1  # Move to the previous character

# Print the reversed string
print("Reversed String:", reversed_string)
````

Q15.  Write a Python program to calculate the factorial of a number provided by the user using a while loop

````python
# Get user input
num = int(input("Enter a number: "))

# Check if the number is negative
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1  # Initialize factorial variable
    i = num  # Start from the given number

    # Calculate factorial using a while loop
    while i > 0:
        factorial *= i  # Multiply the current value
        i -= 1  # Decrease the counter

    # Print the result
    print(f"The factorial of {num} is: {factorial}")
````
